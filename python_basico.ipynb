Estruturas de controle

```python
#Condicionais: if, elif, else
#Loops: for, while

#Estrutura básica do if

nota = float(input('Digite a nota do aluno: '))

if nota >= 7:
    print('Aprovado!')
elif nota >= 5:
    print('Recuperação!')
else:
    print('Reprovado!')

# Operadores de comparação:
# == (igual a)
# != (diferente de)
# > (maior que)
# < (menor que)
# >= (maior ou igual a)
# <= (menor ou igual a)

# Operadores Lógicos:
# and (e)
# or (ou)
# not (não)
```

Loops: for e while

```python
# Loop FOR: percorre uma sequência de elementos
for i in range(5): # range(5) gera os números de 0 a 4
    print(f'Número: {i}')

for chapolin in range(1, 11, 2):
    print(chapolin)

heroi = 'Super Man'
for letra in heroi:
    print(letra)
```

```python
# Loop WHILE: executa um bloco de código enquanto uma condição for verdadeira

tentativas = 0

while tentativas < 3:
    senha = input('Digite a senha: ')
    if senha == 'python123':
        print('Acesso liberado!')
        break

    else:
        print('Senha incorreta. Tente novamente.')
        tentativas += 1
```

##Funções e Módulos em Python

Funções são blocos de código reutilizáveis que recebem dados, processam e devolvem um resultado.

Sâo o bloco de construção fundamental de qualquer programa bem organizado.

```python
# Definindo uma função simples

def saudar(nome):
    mensagem = f'Olá, {nome}!'
    return mensagem

resultado = saudar('Darth Vader')
print(resultado)
```

```python
def calcular_desconto(preco, desconto=10): # desconto padrão de 10%
    valor_desconto = preco * (desconto / 100)
    preco_final = preco - valor_desconto
    return preco_final


print(calcular_desconto(100)) # Usa o desconto padrão de 10%
print(calcular_desconto(300, 25)) # Usa um desconto personalizado de 25%
```

```python
from time import sleep

def calcular_media(nota1, nota2, nota3, nota4):
    media = (nota1 + nota2 + nota3 + nota4) / 4
    return media

def verificar_aprovacao(media):
    if media >= 7:
        return 'Aprovado'
    elif media >= 5:
        return 'Recuperação'
    else:
        return 'Reprovado'

print('Formulário do Aluno')
print('-------------------')
nome = input('Digite seu nome: ')
idade = input('Digite sua idade: ')

nota1 = float(input('Digite sua nota do 1° Bimestre: '))
nota2 = float(input('Digite sua nota do 2° Bimestre: '))
nota3 = float(input('Digite sua nota do 3° Bimestre: '))
nota4 = float(input('Digite sua nota do 4° Bimestre: '))

print('-------------------')
print(f'Olá {nome}, você tem {idade} anos. Vamos calcular sua média final.')
print('Calculando média...')
sleep(2)
print('-------------------')

media_final = calcular_media(nota1, nota2, nota3, nota4)
status = verificar_aprovacao(media_final)

print(f'Sua Média: {media_final:.2f}')
print(f'Status: {status}')
```
