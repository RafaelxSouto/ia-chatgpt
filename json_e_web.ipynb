# JSON

JavaScript Object Notation é um formato de texto usado para representar dados estruturados.

É o formato mais usado para troca de dados entre sistemas e APIs.

Parece muito com dicionário do Python

```python
# JSON é texto  --> parece muito com dicionário em Python

#{'nome': 'Rafael', 'idade': 28, 'ativo': true}

import json

# Converter um dicionário para JSON

dados = {
    "nome": "Rafael",
    "idade": 28,
}

json_string = json.dumps(dados)

print(json_string)
print(type(dados))
print(type(json_string))

# Converter JSON para dicionário
dados_com_endereco = {
    "nome": "Karina",
    "idade": 9.5,
}

text_json = json.dumps(dados_com_endereco, ensure_ascii=False)
print(type(text_json))

meudicionario = json.loads(text_json)
print(meudicionario)

print(meudicionario['nome'])
```

# Consultando uma API pública

Vamos acessar a API do site ViaCep: https://viacep.com.br/

Vamos usar a biblioteca requests do Python (Bilioteca requisição HTTP)

```python
import requests
import json

# 1° Exemplo: Buscar dados de um CEP (ViaCep API)

cep = '04953-1020'
url = f'https://viacep.com.br/ws/{cep}/json/'
try:
    response = requests.get(url)
    print(response.status_code)

    dados = response.json()

    print(dados['logradouro'])
    print(dados['bairro'])
    print(dados['complemento'])
    # print(dados['telefone']) #KeyEroor (não existe a chave no corpo da requisiçãod)
    print(dados['estado'])
except Exception as error:
    print(f'Deu errado ae patrão: {error}')
```

```python
# Código com correções de Erro do Amigão Gemini

import requests

# 1. Entrada de dados e Limpeza (Ignorando o hífen)
entrada = input('Qual seu CEP: ').strip()
cep_limpo = entrada.replace("-", "")

try:
    # 2. Validações de Exceção (Regras de Negócio)
    if not cep_limpo.isdigit():
        raise ValueError("O CEP deve conter apenas números.")

    if len(cep_limpo) != 8:
        raise ValueError("O CEP deve ter exatamente 8 dígitos.")

    # 3. Requisição
    url = f'https://viacep.com.br/ws/{cep_limpo}/json/'
    response = requests.get(url)
    response.raise_for_status() # Verifica se houve erro de conexão (404, 500, etc)

    dados = response.json()

    # 4. Verificação de CEP não encontrado pela API
    if 'erro' in dados:
        raise ValueError("CEP não encontrado na base de dados.")

    # 5. Exibição dos dados
    print('\n' + '-'*10 + ' Sistema de Informação de CEP ' + '-'*10)
    print(f'O CEP solicitado foi: {entrada}')
    print('A informação solicitada foi:')

    # Tratando o complemento (se estiver vazio ou não existir)
    complemento = dados.get('complemento')
    if not complemento: # Se for string vazia ou None
        complemento = "Não informado"

    print(f'Nome da Rua: {dados.get("logradouro", "N/A")}')
    print(f'Complemento: {complemento}')
    print(f'Nome do Bairro: {dados.get("bairro", "N/A")}')
    print(f'Cidade/Estado: {dados.get("localidade", "N/A")} - {dados.get("uf", "N/A")}')

except ValueError as err:
    print(f'Erro de Validação: {err}')
except requests.exceptions.RequestException as err:
    print(f'Erro de Conexão: {err}')
except Exception as err:
    print(f'Erro inesperado: {err}')
```
