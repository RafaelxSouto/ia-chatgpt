```python
# Acessando elementos de uma lista por índice
print (frutas[0]) # Imprime primeiro elemento
print (frutas[1])
print (frutas[-1]) # Imprime 'uva' (último elemento)

# Fatiamento (slicing): pega um pedaço da lista
print (frutas[1, 3]) #
print (frutas[:2]) # Imprime os dois primeiros elementos
```

```python
frutas = [
    'maçã',
    'banana',
    'laranja',
    'uva',
]
# Métodos de lista
print(frutas)
frutas.append('manga') # Adiciona 'manga' ao final da lista
print(frutas)
frutas.insert(1, 'kiwi') # Insere 'kiwi' na posição 1
print(frutas)
frutas.sort() # Ordena a lista em ordem alfabética
print(frutas)
frutas.reverse() # Inverte a ordem da lista
print(frutas)
print(len(frutas)) # Imprime o número de elementos na lista
'banana' in frutas # Verifica se 'banana' está na lista
'kiwi' in frutas
```

## Dicionários

```python
# Dicionários: guardam pares chave:valor

pessoa = {
    'nome': 'Rafael',
    'idade': 24,
    'email': 'rafael@email.com',
    'isAtivo': True
}

# Acessando valores por chave
print(pessoa['nome'])
print(pessoa.get('telefone'))
print(pessoa.get('telefone', 'N/A')) # Valor padrão se chave não existir

# Adicionando valores no divionário

pessoa['telefone'] = '1234-5678'
print(pessoa)
```

```python
# Dicionários aninhados ( muito cuidado em APIs)

usuario = {
    'id': 1,
    'dados': {
        'nome': 'Rafael',
        'endereco': {
            'cidade': 'São Paulo',
            'estado': 'SP'
        }
    }
}
print(usuario['dados']['endereco']['cidade'])
print (usuario)
```

```python
# try:
# except:
# finally:

try :

    numero = int(input('Digite um número: '))
    resultado = 100 / numero
    print(f'Resultado: {resultado}')
except ValueError:
    print('Erro: Você deve digitar um número válido..')
except ZeroDivisionError:
    print('Erro: Não é possível dividir por zero.')
finally:
    print('Execução concluída.')
```

```python
import requests

# Exemplo 1 -- Buscar os dados de um CEP(https://viacep.com.br/)

cep = '01310-100'
url = f'https://viacep.com.br/ws/{cep}/json/'

response = requests.get(url)
print(response.status_code)

dados = response.json()
print(dados['logradouro'])
print(dados)
```
